```python
# Importing libraries.
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.ticker import PercentFormatter
import plotly
import plotly.graph_objects as go
import plotly.express as px
from tqdm import tqdm
```

```python
# Adjust matplotlib.
%matplotlib inline

# set theme
sns.set_theme(style='whitegrid', palette='Set2')

# set plt
plt.rcParams['figure.figsize'] = 10, 5
# plt.rcParams['font.family'] = 'DejaVu Serif'
plt.rcParams['lines.linestyle'] = '-'
plt.rcParams['lines.linewidth'] = 3
plt.rcParams['lines.markersize'] = 8
plt.rcParams['xtick.labelsize'] = 16
plt.rcParams['ytick.labelsize'] = 16
plt.rcParams['legend.fontsize'] = 18
plt.rcParams['axes.titlesize'] = 24
plt.rcParams['axes.labelsize'] = 18
```

```python
data_df = pd.read_csv('data.csv', sep=',', names=['task_name', 'block_size', 'num_elements', 'millis'])
data_df.head()
```

```python
# KernelAdd

task_df = data_df[data_df['task_name'] == 'KernelAdd']

# Worst case
fig, axes = plt.subplots(1, 1, figsize=(12, 7))

# Iterations plot

block_sizes = task_df['block_size'].unique()
cmap = mpl.colormaps['viridis']
colors = [cmap(i) for i in np.linspace(0, 1, len(block_sizes))]

for i, block_size in enumerate(block_sizes):
    plot_data = task_df[task_df['block_size'] == block_size]
    axes.plot(plot_data['num_elements'], plot_data['millis'], label=str(block_size), color=colors[i])

axes.set_yscale('log')
axes.set_title('KernelAdd')
axes.set_xlabel(' ')
axes.set_ylabel(' ()')
axes.legend()

plt.show()
```
