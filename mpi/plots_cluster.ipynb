```python
# Importing libraries.
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.ticker import PercentFormatter
import plotly
import plotly.graph_objects as go
import plotly.express as px

# from tqdm import tqdm
# import jax
# import jax.numpy as jnp
# import torch
```

```python
# Adjust matplotlib.
%matplotlib inline

# set theme
sns.set_theme(style='whitegrid', palette='Set2')

# set plt
plt.rcParams['figure.figsize'] = 10, 5
# plt.rcParams['font.family'] = 'DejaVu Serif'
plt.rcParams['lines.linestyle'] = '-'
plt.rcParams['lines.linewidth'] = 3
plt.rcParams['lines.markersize'] = 8
plt.rcParams['xtick.labelsize'] = 16
plt.rcParams['ytick.labelsize'] = 16
plt.rcParams['legend.fontsize'] = 18
plt.rcParams['axes.titlesize'] = 24
plt.rcParams['axes.labelsize'] = 18
```

```python
# Extract data.
acceleration_df = pd.read_csv('./out/acceleration_cluster.csv', sep=';')

df_3 = acceleration_df[acceleration_df['parts_count'] == 10**3]
df_6 = acceleration_df[acceleration_df['parts_count'] == 10**6]
df_8 = acceleration_df[acceleration_df['parts_count'] == 10**8]

display(acceleration_df.head())
display(df_8.head())
```

```python
# Worst case
fig, axes = plt.subplots(1, 1, figsize=(12, 7))

# Iterations plot
axes.plot(df_3['procs_count'], df_3['acceleration'], label='$10^3$ ')
axes.plot(df_6['procs_count'], df_6['acceleration'], label='$10^6$ ')
axes.plot(df_8['procs_count'], df_8['acceleration'], label='$10^8$ ')

axes.set_title('    ')
axes.set_xlabel(' ')
axes.set_ylabel('')
axes.legend()

plt.show()
```
